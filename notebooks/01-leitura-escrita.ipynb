# Análise de Dados do YouTube com PySpark
## Leitura e Escrita de Arquivos

Neste notebook, iremos trabalhar com dados do YouTube para compreender o comportamento e popularidade de vídeos e influenciadores, utilizando PySpark para leitura e escrita em diferentes formatos.

```python
# Importações necessárias
from pyspark.sql import SparkSession

# Criando a SparkSession
spark = SparkSession.builder \
    .appName('YouTube Analytics') \
    .config('spark.sql.warehouse.dir', '/tmp/spark-warehouse') \
    .enableHiveSupport() \
    .getOrCreate()

print('SparkSession criada com sucesso!')
print(f'Versão do Spark: {spark.version}')
```

## Passo 2 — Leitura do arquivo `videos-stats.csv`

```python
# Passo 2: Leitura do arquivo videos-stats.csv (sem inferência de esquema)
df_videos = spark.read \
    .option('header', 'true') \
    .csv('videos-stats.csv')

print('Arquivo videos-stats.csv lido com sucesso!')
```

## Passo 3 — Visualização dos primeiros 8 registros

```python
# Passo 3: Visualizar os primeiros 8 registros
df_videos.show(8, truncate=False)
```

## Passo 4 — Visualização do esquema

```python
# Passo 4: Visualizar o esquema do arquivo (sem inferência — todos os campos são StringType)
print('Esquema do arquivo videos-stats.csv (sem inferência de esquema):')
df_videos.printSchema()
```

## Passo 5 — Releitura inferindo o esquema

```python
# Passo 5: Reler o arquivo inferindo o esquema automaticamente
df_videos = spark.read \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .csv('videos-stats.csv')

print('Esquema do arquivo videos-stats.csv (COM inferência de esquema):')
df_videos.printSchema()
```

## Passo 6 — Salvar como `videos-parquet` no formato Parquet

```python
# Passo 6: Salvar o arquivo como 'videos-parquet' no formato Parquet com cabeçalho
df_videos.write \
    .option('header', 'true') \
    .mode('overwrite') \
    .parquet('videos-parquet')

print('Arquivo salvo como videos-parquet no formato Parquet com sucesso!')
```

## Passo 7 — Leitura e visualização do arquivo `videos-parquet`

```python
# Passo 7: Ler e visualizar o arquivo 'videos-parquet' com cabeçalho
df_parquet = spark.read \
    .option('header', 'true') \
    .parquet('videos-parquet')

print('Primeiros 8 registros do arquivo videos-parquet:')
df_parquet.show(8, truncate=False)

print('\nEsquema do arquivo videos-parquet:')
df_parquet.printSchema()
```

## Passo 8 — Salvar como tabela `tb_videos` no Spark Catalog

```python
# Passo 8: Salvar como tabela 'tb_videos' no banco de dados default do Spark Catalog
df_parquet.write \
    .mode('overwrite') \
    .saveAsTable('default.tb_videos')

print('Tabela tb_videos salva no banco de dados default do Spark Catalog com sucesso!')
```

## Passo 9 — Listar tabelas do Spark Catalog

```python
# Passo 9: Listar as tabelas do Spark Catalog para verificar a tabela criada
print('Tabelas disponíveis no banco de dados default do Spark Catalog:')
spark.catalog.listTables('default')
```

## Passo 10 — Consulta à tabela `tb_videos` via Spark SQL

```python
# Passo 10: Utilizar Spark SQL para ler a tabela 'tb_videos'
df_sql = spark.sql('SELECT * FROM default.tb_videos')

print('Resultado da consulta SQL na tabela tb_videos:')
df_sql.show(8, truncate=False)

print(f'\nTotal de registros na tabela: {df_sql.count()}')
```

## Passo 11 — Leitura do arquivo `comments.csv` inferindo o esquema

```python
# Passo 11: Ler o arquivo comments.csv inferindo o esquema e visualizar
df_comments = spark.read \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .csv('comments.csv')

print('Primeiros 8 registros do arquivo comments.csv:')
df_comments.show(8, truncate=False)

print('\nEsquema do arquivo comments.csv (com inferência de esquema):')
df_comments.printSchema()
```

## Passo 12 — Salvar como `comments-parquet` no formato Parquet

```python
# Passo 12: Salvar o arquivo como 'comments-parquet' no formato Parquet com cabeçalho
df_comments.write \
    .option('header', 'true') \
    .mode('overwrite') \
    .parquet('comments-parquet')

print('Arquivo salvo como comments-parquet no formato Parquet com sucesso!')

# Verificação: lendo o arquivo salvo
df_comments_parquet = spark.read \
    .option('header', 'true') \
    .parquet('comments-parquet')

print(f'Total de registros em comments-parquet: {df_comments_parquet.count()}')
df_comments_parquet.show(5, truncate=False)
```

## Encerramento da SparkSession

```python
# Encerrar a SparkSession ao finalizar
spark.stop()
print('SparkSession encerrada com sucesso!')
```
